```python
import os
import matplotlib.pyplot as plt
import cv2
```

```python
BASE_DIR = "../data/soil_images"
TRAIN_DIR= os.path.join(BASE_DIR,"train_data")
TEST_DIR=os.path.join(BASE_DIR,"test")
```

```python
soil_types_train = os.listdir(TRAIN_DIR)
soil_types_train
```

```python
soil_types_test = os.listdir(TEST_DIR)
soil_types_test
```

```python
for soil in soil_types_train:
    count=len(os.listdir(os.path.join(TRAIN_DIR,soil)))
    print(f"{soil}:{count}")
```

```python
for soil in soil_types_test:
    count=len(os.listdir(os.path.join(TEST_DIR,soil)))
    print(f"{soil}:{count}")
```

```python
soil = soil_types_train[3]
img_name = os.listdir(os.path.join(TRAIN_DIR, soil))[0]
img_path = os.path.join(TRAIN_DIR, soil, img_name)

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title(f"Train - {soil}")
plt.axis("off")
```

```python
soil = soil_types_test[3]
img_name = os.listdir(os.path.join(TEST_DIR, soil))[0]
img_path = os.path.join(TEST_DIR, soil, img_name)

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title(f"Test - {soil}")
plt.axis("off")
```

```python
plt.figure(figsize=(8,6))
index=1

for soil in soil_types_train:
    img_name=os.listdir(os.path.join(TRAIN_DIR,soil))[0]
    img_path=os.path.join(TRAIN_DIR,soil,img_name)
    img=cv2.imread(img_path)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.subplot(2,2,index)
    plt.imshow(img)
    plt.title(soil)
    plt.axis("off")
    index+=1

plt.tight_layout()
plt.show()
```

### Conclusion
- Soil images are organized into train and test sets
- Four soil types are present in both sets
- Images are readable and correctly labeled
- Dataset is suitable for CNN training
